```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, exp, sin, cos
```

```python
# Part 1: Compute the first 4 terms of the Taylor Series for e^x around x=0
def taylor_exp(x, n_terms):
    result = 0
    for n in range(n_terms):
        result += (x**n) / factorial(n)
    return result

# Test values
test_values = [-2, -1, 0, 1, 2]
print("Part 1: Taylor Series for e^x around x=0")
print("-" * 50)
print("x\tApproximation (4 terms)\tActual Value\tAbsolute Error")
for x in test_values:
    approx = taylor_exp(x, 4)
    actual = exp(x)
    error = abs(approx - actual)
    print(f"{x}\t{approx:.6f}\t\t{actual:.6f}\t{error:.6f}")

plt.show()
```

```python
# Part 2: Taylor Series for sin(x)
def taylor_sin(x, n_terms):
    result = 0
    for n in range(n_terms):
        term = ((-1)**n) * (x**(2*n+1)) / factorial(2*n+1)
        result += term
    return result

# Create plot for sin(x) and its approximations
x_values = np.linspace(-2*np.pi, 2*np.pi, 1000)
actual_sin = np.sin(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, actual_sin, 'b-', label='sin(x)')

for terms in [1, 3, 5]:
    approx_sin = [taylor_sin(x, terms) for x in x_values]
    plt.plot(x_values, approx_sin, label=f'Taylor sin(x) ({terms} terms)')

plt.title('sin(x) and Taylor Series Approximations')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.savefig('sin_taylor_approximations.png')

plt.show()
```

```python
# Part 3: Compute and plot the absolute error for sin(x) Taylor approximation with 5 terms
approx_sin_5 = [taylor_sin(x, 5) for x in x_values]
error_sin_5 = [abs(a - s) for a, s in zip(approx_sin_5, actual_sin)]

plt.figure(figsize=(10, 6))
plt.plot(x_values, error_sin_5)
plt.title('Absolute Error of sin(x) Taylor Approximation (5 terms)')
plt.xlabel('x')
plt.ylabel('Absolute Error')
plt.yscale('log')  # Using log scale to better visualize the error
plt.grid(True)
plt.savefig('sin_taylor_error.png')

# Taylor Series for cos(x)
def taylor_cos(x, n_terms):
    """
    Compute the Taylor series approximation of cos(x) around x=0 with n_terms
    """
    result = 0
    for n in range(n_terms):
        term = ((-1)**n) * (x**(2*n)) / factorial(2*n)
        result += term
    return result

# Create plot for cos(x) and its 4-term approximation
actual_cos = np.cos(x_values)
approx_cos_4 = [taylor_cos(x, 4) for x in x_values]

plt.figure(figsize=(10, 6))
plt.plot(x_values, actual_cos, 'b-', label='cos(x)')
plt.plot(x_values, approx_cos_4, 'r--', label='Taylor cos(x) (4 terms)')
plt.title('cos(x) and Taylor Series Approximation (4 terms)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.savefig('cos_taylor_approximation.png')

plt.show()
```
